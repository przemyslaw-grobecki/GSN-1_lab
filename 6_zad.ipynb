```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a linearly separable dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=42)

# Create a scatter plot of the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Linearly Separable Dataset')
plt.show()

# Create a single-neuron network
model = Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid', use_bias=False))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=100, verbose=0)

# Plot the decision boundary
x1 = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
x2 = -(model.get_weights()[0][0][0] / model.get_weights()[0][1][0]) * x1
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.plot(x1, x2, c='r')
plt.title('Linearly Separable Dataset with Decision Boundary')
plt.show()
```
